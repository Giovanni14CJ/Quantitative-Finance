```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
precios = pd.read_excel('Ruta de archivo.xlsx', index_col = 'Date')
```

```python
rendimientos = (precios - precios.shift(1)) / precios.shift(1)
rendimientos
```

```python
numeroactivos = len(rendimientos.columns)
numeroactivos
```

```python
rportafolio = []
sdportafolio = []
pesosportafolio = []
```

```python
for x in range(5000):
    pesos = np.random.random(numeroactivos)
    pesos /= np.sum(pesos)
    pesosportafolio.append(pesos)
    rportafolio.append(np.dot(rendimientos.mean(), pesos))
    sdportafolio.append(np.sqrt(np.dot(pesos.T, np.dot(rendimientos.cov(), pesos))))
    
```

```python
pesosportafolio
```

```python
rportafolio
```

```python
# Ahora crearemos un diccionario con el rendimiento, volatilidad y pesos de los 20 activos para cada uno de los 5000 portafolios
```

```python
diccionario = {'Rendimientos':rportafolio, 'Volatilidad':sdportafolio}
for contador, ticker in enumerate(rendimientos.columns.tolist()):
    diccionario['Peso ' + ticker] = [w[contador] for w in pesosportafolio]
```

```python
matrizportafolios = pd.DataFrame(diccionario)
matrizportafolios
```

```python
# Frontera Eficiente
matrizportafolios.plot(x = 'Volatilidad', y = 'Rendimientos', kind = 'scatter')
```

```python
varianzaminima = matrizportafolios.iloc[matrizportafolios['Volatilidad'].idxmin()]
varianzaminima
```

```python
matrizportafolios.plot(x = 'Volatilidad', y = 'Rendimientos', kind = 'scatter')
plt.scatter(x = varianzaminima[1], y = varianzaminima[0], color = 'red', marker = '*', s = 200)
```

```python
# Portafolio optimo
rf = 0.0165 / 12
optimo = matrizportafolios.iloc[((matrizportafolios['Rendimientos'] - rf) / matrizportafolios['Volatilidad']).idxmax()]
optimo
```

```python
matrizportafolios.plot(x = 'Volatilidad', y = 'Rendimientos', kind = 'scatter')
plt.scatter(x = varianzaminima[1], y = varianzaminima[0], color = 'red', marker = '*', s = 200)
plt.scatter(x = optimo[1], y = optimo[0], color = 'black', marker = '*', s = 200)
```
